```python
import numpy as np
from numpy import pi, sin, cos, sqrt
import matplotlib.pyplot as plt

width = 200
height = 200
img = np.zeros((width, height))

for y in range(height):
  phase_y = y / height
  for x in range(width):
    phase_x = x / width
    distance = sqrt(phase_x**2+phase_y**2)
    img[x, y] = distance

plt.imshow(img, cmap='gray')
plt.show()
```
